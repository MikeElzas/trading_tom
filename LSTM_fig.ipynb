```python
import matplotlib.pyplot as plt
```

```python
from model.lstm_model import lstm_model_return
```

```python
dict = lstm_model_return(["BTC/USDT", "ETH/USDT", "SOL/USDT"])
```

```python
history = dict["BTC/USDT"][1]
```

```python
# overfitting or underfitting analysis based on epochs
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
y_test_inverse2 = dict["BTC/USDT"][3]
y_hat_inverse2 = dict["BTC/USDT"][4]
```

```python
plt.plot(y_test_inverse2, label="Actual Price", color='green')
plt.plot(y_hat_inverse2, label="Predicted Price", color='red')
 
plt.title('Bitcoin price prediction')
plt.xlabel('Time [days]')
plt.ylabel('Price')
plt.legend(loc='best')
 
plt.show();
```
